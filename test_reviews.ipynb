```python
import joblib

model = joblib.load('models/logistic_model.joblib')
vectorizer = joblib.load('models/vectorizer.joblib')

reviews = [
    "This was the best movie I've ever seen!",
    "Terrible waste of time. I hated the characters.",
    "It was okay, not great but not the worst either."
]

reviews_vec = vectorizer.transform(reviews)
predictions = model.predict(reviews_vec)

label_map = {0: 'NEGATIVE', 1: 'POSITIVE'}
for review, pred in zip(reviews, predictions):
    print(f"[{label_map[pred]}] -> {review}")
```
